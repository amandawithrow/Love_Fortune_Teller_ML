```python
# EDUCATION GAP BIN
edu_gap_bin_df = df.groupby(['edu_gap_bin'])['caseid_new'].count()
edu_gap_bin_df = edu_gap_bin_df.reset_index()
edu_gap_bin_df['Percent'] = round((edu_gap_bin_df['caseid_new']/3009*100),2)


edu_gap_bin_dict = {}

for i in range(len(edu_gap_bin_df['edu_gap_bin'])):
    #edu_gap_bin_df['edu_gap_bin'][i] = iloc(i,0)
    edu_gap_bin_dict[edu_gap_bin_df['edu_gap_bin'][i]] = edu_gap_bin_df['Percent'][i]
print(edu_gap_bin_dict)
edu_gap_bin_df
```

```python
how_met_unique_df = df.groupby(['how_met_unique'])['caseid_new'].count()
how_met_unique_df = how_met_unique_df.reset_index()
how_met_unique_df['Percent'] = round((how_met_unique_df['caseid_new']/3009*100),2)


how_met_unique_dict = {}
for i in range(len(how_met_unique_df['how_met_unique'])):
    #edu_gap_bin_df['edu_gap_bin'][i] = iloc(i,0)
    how_met_unique_dict[how_met_unique_df['how_met_unique'][i]] = how_met_unique_df['Percent'][i]
print(how_met_unique_dict)
how_met_unique_df
```
